```
def get_producers_directors(string_rep_crew_col):
  
  crew_list = literal_eval(string_rep_crew_col)
  
  producer_list =[]
  director_list =[]
  
  for elem in crew_list:
    if elem['job']== 'Producer':
      producer_list.append(elem['name'])
    if elem['job']== 'Director':
      director_list.append(elem['name'])

  return producer_list, director_list
```

```
get_producers_directors(no_nan_crew.loc[1,'crew'])
```

```
#to ignore nan errors we can use the try except block for using it with the original df containing all the movies
#where we want to apply the func #so if there's nan values we can restrict it to return empty lists instead of throwing errors!

def get_producers_directors(string_rep_crew_col):

  try:
  
    crew_list = literal_eval(string_rep_crew_col)
    
    producer_list =[]
    director_list =[]
    
    for elem in crew_list:
      if elem['job']== 'Producer':
        producer_list.append(elem['name'])
      if elem['job']== 'Director':
        director_list.append(elem['name'])

    return producer_list, director_list

  except:
    return [],[]
```

```
df['Prod&Direc'] = df.apply(lambda x: get_producers_directors(x['crew']),axis=1)
```

```
df.head()
```

```
#but we have to convert the tuple containing two list in Prod&Direc col in df to seperate cols having producers & directors.

df[['Producers','Directors']] = df.apply(lambda x: get_producers_directors(x['crew']),axis=1,result_type='expand')
```

```
df.head()
```

```
for index, row in df.iterrows():
  print(row['Producers'])
  if index>4:
    break
```

```
for index,row in df.iterrows():
  for producer in row['Producers']:
    print(producer)
  if index>5:
    break
```

```
producer_dict ={}
for index,row in df.iterrows():
  # print(row['ROI'])
  for producer in row['Producers']:
    # print(producer)
    if producer not in producer_dict.keys():
      #add producer in dict
      producer_dict.update({producer:[row['ROI']]})
      # producer_dict[producer]=[row['ROI']]
    else:
      #update the ROI for this producer
      value= producer_dict[producer]
      # print(value)
      value.append(row['ROI'])
      producer_dict[producer]=value
```

```
producer_dict
```

```
#we dont have equal no of values for each producers, since some producers have multiple movies
#we want the keys from dictionary to be rows and not cols 
pd.DataFrame.from_dict(producer_dict, orient='index')
```

```
pd.DataFrame.from_dict(producer_dict, orient='index').mean(axis=1).reset_index().rename(columns={'index':'Producers',0:'Avg_roi'})
```

```
#to get the top 3producers!
pd.DataFrame.from_dict(producer_dict, orient='index').mean(axis=1).reset_index().rename(columns={'index':'Producers',0:'Avg_roi'}).sort_values('Avg_roi',ascending=False)
```

```
def actor_func(cast):

  try:

    actor =[]

    cast_lit = literal_eval(cast)

    for ele in cast_lit:
      actor.append(ele['name'])

    return actor

  except:

    return []
```

```
df['actors'] = df.apply(lambda r: actor_func(r['cast']),axis=1)
```

```
#lets create a dict to create a mapping of actors with its particular movie!

actors_dict = {}

for index,actr in df.iterrows():
  for items in actr['actors']:
    if items not in actors_dict.keys():
      actors_dict.update({items:[actr['original_title']]})
      
    else:
      value1= actors_dict[items]
      value1.append(actr['original_title'])
      actors_dict[items] = value1
```
